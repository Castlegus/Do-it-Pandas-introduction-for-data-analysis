```python
# groupby 메서드
import pandas as pd
df = pd.read_csv('C:/Users/정성현/Do it! 데이터 분석을 위한 판다스 입문/data/gapminder.tsv', sep = '\t')
```

```python
# 사용자 함수와 groupby 메서드를 조합하려면 agg 메서드를 이용한다.
def my_mean(values):
    n = len(values)
    sum = 0
    for value in values:
        sum += value
    return sum/n

df.groupby('year')['lifeExp'].agg(my_mean) # 그런데 apply 함수랑 뭐가 다른거지?
```

```python
# 2개의 인자값을 받아 처리하는 사용자 함수. 평균값을 구하여 diff_value에 입력한 값을 뺀다.
def my_mean_diff(values, diff_value):
    n = len(values)
    sum = 0
    for value in values:
        sum += value
    mean =  sum/n
    return mean - diff_value

global_mean = df.lifeExp.mean()
df.groupby('year').lifeExp.agg(my_mean_diff, diff_value = global_mean)
```

```python
# 여러 개의 집계 매서드 한 번에 사용하기
import numpy as np
df.groupby('year')['lifeExp'].agg([np.count_nonzero, np.mean, np.std])
```

```python
df.groupby('year').agg({'lifeExp':'mean','pop':'median','gdpPercap':'median'})
```

```python
# 표준 점수 계산하기
def my_zscore(x):
    return (x-x.mean()) / x.std()

df.groupby('year')['lifeExp'].transform(my_zscore) # transform은 또 apply와 뭐가 다를까..?
```

```python
# 누락 값을 평균 값으로 처리하기

import seaborn as sns
import numpy as np

np.random.seed(42)
tips_10 = sns.load_dataset('tips').sample(10)
tips_10.loc[np.random.permutation(tips_10.index)[:4], 'total_bill'] = np.NaN #permutation은 shuffle과 뭐가 다를까..?
tips_10
```

```python
tips_10.groupby('sex').count()
```

```python
# 성별을 구분하여 total_bill 열의 데이터를 받아 평균 값을 구하는 함수 만들기
def fill_na_mean(x):
    avg = x.mean()
    return x.fillna(avg)
```

```python
# NaN 값을 성별을 구분한 평균 값으로 채운다.
total_bill_group_mean = tips_10.groupby('sex').total_bill.transform(fill_na_mean)
tips_10['fill_total_bill'] = total_bill_group_mean
tips_10
```

```python
# 데이터 필터링 filter 메소드
tips = sns.load_dataset('tips')
tips
```

```python
tips['size'].value_counts()
```

```python
# 1번, 5번, 6번 테이블의 주문이 매우 적다.
# 따라서 30번 이상 주문이 있는 테이블만 분석을 하려고 한다.
tips_filtered = tips.groupby('size').filter(lambda x: x['size'].count()>30)
tips_filtered
```
